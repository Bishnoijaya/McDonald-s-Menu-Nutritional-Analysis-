# McDonald's Menu Nutritional Analysis


# Problem Statement
McDonald's is a global fast-food chain known for its diverse menu offerings. As a data analyst, your task is to analyze the nutritional content of the menu items available at McDonald's outlets. This analysis will provide valuable insights into the calorie count and nutrition facts of various menu items.

 ###### Items = Items name displayed on menu

 ###### Total Fat = Total Fat of an item
 ###### Total Fat(% Daily Value) = Daily percentage value of Fat

 ###### Dietar = Total Dietary fiber in an menu item
 ###### Sugars = Totsl Sugars count in an item
 ###### Protein = Total Protein in the menu Item
 ###### Vitamin A(% Daily Value) = Daily percentage value of Vitamin A
 ###### Vitamin c(% Daily Value) = Daily percentage value of Vitamin C
 ###### Calcium(% Daily Value) = Daily percentage value of Calcium

# Outlier treatment 

```python
col=MCD2.select_dtypes(exclude='object').columns
Q1 = MCD2[col].quantile(0.25)
Q3 = MCD2[col].quantile(0.75)
IQR = Q3 - Q1
MCD2= MCD2[~((MCD2[col] < (Q1 - 1.5*IQR))| (MCD2[col] > (Q3 + 1.5*IQR))).any(axis=1)]
```
